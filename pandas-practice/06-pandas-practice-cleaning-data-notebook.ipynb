## Handling Missing Values and Renaming Columns

This notebook demonstrates basic data cleaning operations in Pandas, including handling missing values and renaming columns.  
These steps are commonly performed to prepare data for analysis.

---

## Concepts Covered

- Identifying and handling missing values
- Filling missing values in specific columns
- Modifying the DataFrame in place
- Renaming columns for clarity and consistency

---

## Examples Used in This Notebook

- Identify rows with missing values using `dropna`
- Fill missing values in the `OrderId` column with a default value
- Use `inplace=True` to apply changes directly to the original DataFrame
- Rename a column to improve readability


```python
import pandas as pd
df =  pd.read_csv('orders.csv')
```

```python
df.dropna  #drops all rows with any missing values
```

```python
df.fillna({'OrderId': 0}, inplace=True)  #fills missing values in 'OrderId' column with 0
# inplace=True modifies the original dataframe, otherwise it returns a new dataframe with the changes
```

```python
df.rename(columns={'OrderID': 'Order ID'}, inplace=True)  #renames 'OrderId' column to 'Order ID'
# inplace=True modifies the original dataframe
```
