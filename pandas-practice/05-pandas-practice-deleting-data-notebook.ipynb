## Removing Rows in Pandas

Pandas allows rows to be removed from a DataFrame using the `drop()` method.  
This notebook demonstrates how to delete an individual row by its index label and verify that the deletion was successful.

---

## Concepts Covered

- Removing a single row using `drop()`
- Understanding that `drop()` works on index labels
- Verifying row deletion using `tail()`

---

## Example Used in This Notebook

- Delete the row with index label `39`
- Display the last 5 rows of the DataFrame to confirm the row was removed


```python
import pandas as pd
df =  pd.read_csv('orders.csv')
```

```python
df = df.drop(39)  #for removing individual row
```

```python
df.tail()   #display last 5 rows of the dataframe to check if row 39 is deleted
```
