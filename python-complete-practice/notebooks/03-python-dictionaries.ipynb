## Purpose

Dictionaries are essential for structured data storage and fast lookups in Python.


```python
dict = {'k1': 'cat', 'k2': 'dog', 'k3': 'mouse', 'k4': 'fish'}
dict
```

```python
dict ['k5'] = 'parrot'
dict
```

```python
dict['k2'] = 'squirrel'
dict
```

```python
dep_workers = {'dep_1': 'Peter', 'dep2': ['Jennifer', 'Michael', 'Tommy']}
dep_workers['dep2']
```

```python
Team = {}
Team['Point Guard'] = 'Dirk'
Team['Shooting Guard'] =  'AI'
Team['Small Forward'] = 'Sean'
Team['Power Forwards'] = 'Alexander'
Team['Center'] = 'Hector'
```

```python
print(Team)
```

```python
print(Team ['Center'])
```

```python
print(Team.get('Small Forward'))
```

```python
print(Team.get('Coach'))
```
