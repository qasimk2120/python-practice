## Concepts Covered

---

## Purpose

Lambda functions are commonly used for concise operations, especially in functional-style programming.


```python
raise_to_power_two = lambda x: x**2 
print(raise_to_power_two(3))
```

```python
(lambda x: x / 2) (11)
```

```python
(lambda x: (2 + 5 * x ** 4) ** 2 / (x + 3) ** 3) (2)
```

```python
sum_xy =  lambda x,y : x + y
sum_xy(2,3)
```

```python
sum_xy(2, lambda x: x*5)
```

```python
(lambda x: (135 - x ** 3) ** 4 / (1 + x)** 5 ) (3)
```
