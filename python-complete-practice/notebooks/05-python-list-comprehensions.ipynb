# Python List Comprehensions

## Overview

This notebook demonstrates list comprehensions as a concise and Pythonic way to create lists.

---

## Concepts Covered

- Basic list comprehension syntax
- Conditional logic inside comprehensions
- Comparison with traditional `for` loops

---

## Purpose

List comprehensions improve readability and performance when working with lists.


```python
numbers = [1, 13, 4, 5, 63, 100]
numbers
```

```python
new_numbers = []
for n in numbers:
    new_numbers.append(n * 2)   #An output expression for an element in iterable #iterable - the sequence we are going to iterate over
print(new_numbers)
#list comprehensions can provide a consise way to build a new list from a given list
```

```python
new_numbers = [n * 2 for n in numbers]
print(new_numbers)
```

```python
for i in range(2):
    for j in range(5):
        print(i + j, end = ' ')
```

```python
new_list_comprehension_1 = [i +j for i in range(2) for j in range(5)]
new_list_comprehension_1
```

```python
type(new_list_comprehension_1)
```

```python
print(new_list_comprehension_1)
```

```python
new_list_comprehension_2 = [[ i + j for i in range(2)] for j in range(5)]
new_list_comprehension_2
```

```python
type(new_list_comprehension_2)
```

```python
type(new_list_comprehension_2[1])
```

```python
type(new_list_comprehension_1[1])
```

```python
list(range(1, 11))
```

```python
[num ** 3 for num in range(1, 11) if num % 2 != 0]
```

```python
for i in range(1, 11):
    if i % 2 != 0:
        print(i ** 3 , end = "  " )
```

```python
[num ** 3 if num % 2 != 0 else "even" for num in range(1, 11)]
```
