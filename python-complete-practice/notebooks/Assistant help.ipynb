```python
def count_values_less_than_20(input_list):
    count = 0
    for value in input_list:
        if float(value) < 20:
            count += 1
    return count

test_list1 = [5, 25, 10, 30, 15, 35]
result = count_values_less_than_20(test_list1)
print(result)
```

```python
prices = {
    "box_of_spaghetti" : 4,
    "lasagna" : 5,
    "hamburger" : 2
}
quantity = {
        "box_of_spaghetti" : 6,
    "lasagna" : 10,
    "hamburger" : 2
}
money_spent = 0
for i in prices:
    money_spent = money_spent + (prices[i] * quantity[i])
print(money_spent)
```
